```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def scraping_eng(query):
    url = f'https://tr-ex.me/%E0%B8%81%E0%B8%B2%E0%B8%A3%E0%B9%81%E0%B8%9B%E0%B8%A5/%E0%B8%AD%E0%B8%B1%E0%B8%87%E0%B8%81%E0%B8%A4%E0%B8%A9-%E0%B9%84%E0%B8%97%E0%B8%A2/{query}'
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Create empty lists to store the scraped data
    name_bolds_eng = []
    name_bolds_thai = []
    all_content_eng = []
    all_content_thai = []

    # Scrape the data and populate the lists
    elements = soup.find_all("div", {'class': 'ctx'})
    count = 1
    for element in elements:
        # Eng 
        try:
            names = element.find("div", {'class': 'stc stc-s'}).find("div",{'class','ctx-text'})
            bolds = names.find_all("a")
            nameBoldsEng = [bold.text for bold in bolds]
            name_bolds_eng.append(nameBoldsEng)
            if names is not None:
                all_content_eng.append(names.text)
        except:
            pass
        # Thai
        try:
            names = element.find("div", {'class': 'stc stc-t ba'}).find("div",{'class','ctx-text'})
            bolds = names.find_all("a")
            nameBoldsThai = [bold.text for bold in bolds]
            name_bolds_thai.append(nameBoldsThai)
            if names is not None:
                all_content_thai.append(names.text)
        except:
            pass

    # Create a dictionary from the scraped data
    data = {
        'nameEng': query,
        'nameBoldEng': name_bolds_eng,
        'nameBoldThai': name_bolds_thai,
        'contentsEng': all_content_eng,
        'contentsThai': all_content_thai,
    }

    return data

query = "anesthesia machine"
df_name_eng = scraping_eng(query)
df = pd.DataFrame(df_name_eng)
df
```
